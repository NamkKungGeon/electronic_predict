### 모듈 불러오기

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import spearmanr
```

### 데이터 불러오기

```python
df_2013 = pd.read_csv("C:/Users/gun67/전력 프로젝트/preprocessing_data/2013_preprocessing_data", encoding = "utf-8-sig")
df_2014 = pd.read_csv("C:/Users/gun67/전력 프로젝트/preprocessing_data/2014_preprocessing_data", encoding = "utf-8-sig")
df_2015 = pd.read_csv("C:/Users/gun67/전력 프로젝트/preprocessing_data/2015_preprocessing_data", encoding = "utf-8-sig")
df_2016 = pd.read_csv("C:/Users/gun67/전력 프로젝트/preprocessing_data/2016_preprocessing_data", encoding = "utf-8-sig")
```

### 상관관계 분석

```python
#2013
# 상관계수 계산
correlation_matrix = df_2013[["Watt", "Temp('C)", "Humidity(%)"]].corr(method = "pearson")
print("\n상관 행렬:\n", correlation_matrix)

# 히트맵 시각화
plt.figure(figsize = (10, 8))
sns.heatmap(correlation_matrix, annot = True, cmap = "coolwarm", vmin = 1, vmax = 1)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#2014
# 상관계수 계산
correlation_matrix = df_2014[["Watt", "Temp('C)", "Humidity(%)"]].corr(method = "pearson")
print("\n상관 행렬:\n", correlation_matrix)

# 히트맵 시각화
plt.figure(figsize = (10, 8))
sns.heatmap(correlation_matrix, annot = True, cmap = "coolwarm", vmin = 1, vmax = 1)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
